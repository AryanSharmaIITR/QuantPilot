```python
import pandas as pd
```

```python
df=pd.read_csv("../data/preprocessed/stock_data_stacked.csv")
df.head()
```

```python
print(df.head())
```

```python
df = df.sort_values("Date")
```

```python
df["stock_id"] = df["stock_name"].astype("category").cat.codes
df = df.drop(columns=["stock_name"])
```

```python
df
```

```python
df.dropna(inplace=True)
```

```python
df["Target"]=df["Target"].apply(lambda x: 1 if x > 0 else 0)
```

```python
df["Date"] = pd.to_datetime(df["Date"])

train = df[(df["Date"] >= "2023-04-03") & (df["Date"] <= "2024-12-31")]
val   = df[(df["Date"] >= "2025-01-01") & (df["Date"] <= "2025-06-30")]
test  = df[(df["Date"] >= "2025-07-01") & (df["Date"] <= "2026-04-02")]
```

```python
features = [col for col in df.columns if col not in ["Date", "Target"]]

X_train, y_train = train[features], train["Target"]
X_val, y_val     = val[features], val["Target"]
X_test, y_test   = test[features], test["Target"]
```

```python
X_train.columns
```

```python
from xgboost import XGBClassifier

model = XGBClassifier(
    n_estimators=500,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42
)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=True
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = model.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, predictions, target_names=["Down/No Gain", "Up"]))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, predictions))
```

```python
model = XGBClassifier(
    n_estimators=500,
    max_depth=5,
    learning_rate=0.05,
    subsample=0.8,
    colsample_bytree=0.8,
    random_state=42,
    scale_pos_weight=1.36
)

model.fit(
    X_train, y_train,
    eval_set=[(X_val, y_val)],
    verbose=True
)
```

```python
from sklearn.metrics import classification_report, confusion_matrix

predictions = model.predict(X_test)

print("Classification Report:\n")
print(classification_report(y_test, predictions, target_names=["Down/No Gain", "Up"]))

print("\nConfusion Matrix:\n")
print(confusion_matrix(y_test, predictions))
```
